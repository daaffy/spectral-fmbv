```python
import numpy as np
import matplotlib.pyplot as plt

f = open("2.txt", "r")

ftxt =  f.read()
print(ftxt)

def find(s, ch):
    return [i for i, ltr in enumerate(s) if ltr == ch]

find(ftxt, ':')
```

```python
data = ftxt[148146:-2]
data = data.split(',')
int_data = [int(curr) for curr in data]
print(int_data)
print(len(int_data))
```

```python
a = 256
b = 469
print(a*b)


img = np.zeros((a,b))
for i in range(a):
    for j in range(b):
        img[i,j] = int_data[j + b*i]
```

```python
plt.imshow(img)
```
